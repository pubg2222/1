```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('EduData.csv') 
```

```python
data
```

```python
data.describe()
```

```python
data = data.replace("?", np.NaN)
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data["Topic"].value_counts()
```

```python
data['Topic'].replace(np.NaN, 'IT', inplace=True)
```

```python
data['Topic'].isnull().sum()
```

```python
data = data.dropna(subset = ['Relation','GradeID'])
```

```python
data.isnull().sum()
```

```python
avg = data['raisedhands'].astype('float').mean()
avg
```

```python
data["raisedhands"].replace(np.NaN, 46, inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Semester'].value_counts()
```

```python
data = data[data['Semester'].str.contains("T") == False]
```

```python
data.reset_index(drop = True, inplace = True)
```

```python
data
```

```python
data.drop(['ParentAnsweringSurvey'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['raisedhands'] =  data['raisedhands'].astype('float') / data['raisedhands'].astype('float').max()
```

```python
data['VisITedResources'] =  data['VisITedResources'] / data['VisITedResources'].astype('float').max()
```

```python
data.head()
```

```python
data['GradeID'].value_counts()
```

```python
data['GradeID'].replace("G-02", 1, inplace = True)
data['GradeID'].replace("G-04", 2, inplace = True)
data['GradeID'].replace("G-05", 3, inplace = True)
data['GradeID'].replace("G-06", 4, inplace = True)
data['GradeID'].replace("G-07", 5, inplace = True)
data['GradeID'].replace("G-08", 6, inplace = True)
data['GradeID'].replace("G-09", 7, inplace = True)
data['GradeID'].replace("G-10", 8, inplace = True)
data['GradeID'].replace("G-11", 9, inplace = True)
data['GradeID'].replace("G-12", 10, inplace = True)
```

```python
data['StudentAbsenceDays'].value_counts()
```

```python
data['StudentAbsenceDays'].replace("Under-7", 0, inplace = True)
data['StudentAbsenceDays'].replace("Above-7", 1, inplace = True)
```

```python
bins = np.linspace(min(data['Discussion']), max(data['Discussion']), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
data['Discussion-Binned'] = pd.cut(data['Discussion'], bins, labels = group_names, include_lowest = True)
```

```python
data[['Discussion','Discussion-Binned']].head()
```

```python
data.to_csv("DW2.csv")
```
