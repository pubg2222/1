```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("online_store_customer_data.csv")
```

```python
#sorting column by
df = data.sort_values(by=['Age'], ascending=True)
df
```

```python
df['Age'].value_counts()
```

```python
df.dtypes
```

```python
agg_functions = {'Amount_spent':['mean', 'median', 'min', 'max', 'std']}
sumStat = df.groupby(['Age']).agg(agg_functions)
```

```python
sumStat
```

```python
categoricalList = df.groupby(by = 'Age').agg('Amount_spent').sum().round(2).tolist()
```

```python
ageList =  df['Age'].unique().tolist()
```

```python
Cat_List = dict(zip(ageList,categoricalList))
```
